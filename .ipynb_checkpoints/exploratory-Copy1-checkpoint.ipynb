```python
# This script performs an exploratory data analysis of the passengers


# VARIABLE DESCRIPTIONS:
# 1  survival        Survival
#                    (0 = No; 1 = Yes)
# 2  pclass          Passenger Class
#                    (1 = 1st; 2 = 2nd; 3 = 3rd)
# 3  name            Name
# 4  sex             Sex
# 5  age             Age
# 6  sibsp           Number of Siblings/Spouses Aboard
# 7  parch           Number of Parents/Children Aboard
# 8  ticket          Ticket Number
# 9  fare            Passenger Fare
# 10 cabin           Cabin
# 11 embarked        Port of Embarkation
#                    (C = Cherbourg; Q = Queenstown; S = Southampton)

# SPECIAL NOTES:
# Pclass is a proxy for socio-economic status (SES)
#  1st ~ Upper; 2nd ~ Middle; 3rd ~ Lower
# 
# Age is in Years; Fractional if Age less than One (1)
#  If the Age is Estimated, it is in the form xx.5
# 
# With respect to the family relation variables (i.e. sibsp and parch)
# some relations were ignored.  The following are the definitions used
# for sibsp and parch.
# 
# Sibling:  Brother, Sister, Stepbrother, or Stepsister of Passenger Aboard Titanic
# Spouse:   Husband or Wife of Passenger Aboard Titanic (Mistresses and Fiances Ignored)
# Parent:   Mother or Father of Passenger Aboard Titanic
# Child:    Son, Daughter, Stepson, or Stepdaughter of Passenger Aboard Titanic
# 
# Other family relatives excluded from this study include cousins,
# nephews/nieces, aunts/uncles, and in-laws.  Some children travelled
# only with a nanny, therefore parch=0 for them.  As well, some
# travelled with very close friends or neighbors in a village, however,
# the definitions do not support such relations.
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import csv as csv
```

```python
csv_file_object = csv.reader(open('train.csv'))       # Load in the csv file
csv_file_object
header = next(csv_file_object)                        # Skip the fist line as it is a header
data=[]                                               # Create a variable to hold the data

for row in csv_file_object:                           # Skip through each row in the csv file,
    data.append(row[0:])                              # adding each row to the data variable
data = np.array(data)                                 # Then convert from a list to an array.
```

```python
print(data[0:10,:])
```

```python
# Compute some important variables
n_passengers=np.size(data[:,1])
n_survivors=np.sum(data[:,1].astype(np.int))

print('N. of passengers:',n_passengers, ' || N. of survivors:',n_survivors,
      ' || % of survivors:',100*n_survivors/n_passengers)
```

```python
# Define data labels
data_label=np.array(['id','survival','pclass','name','sex','age','sibsp','parch','ticket','fare','cabin','embarked'])

# Check how much data is available
data_available=np.zeros((1,12))
for x in range(0,12):
    data_available[0,x]=np.size(data[:,x].nonzero())
    print(data_label[x], ':',data_available[0,x],'of',n_passengers,'--->',100*data_available[0,x]/n_passengers,'%')

# Plot fraction of available data
plt.figure(1, figsize=(13,6))
n_var = 12

index = np.arange(n_var)
bar_width = 0.8

opacity = 0.4

rects1 = plt.bar(index, 100*np.transpose(data_available)/n_passengers, bar_width,
                 alpha=opacity,
                 color='b')

plt.axis([-.5, 12, 0, 110])
plt.xlabel('variable')
plt.ylabel('Data available')
plt.title('Available data for each variable')
plt.xticks(index + bar_width/2, data_label)

plt.grid(True)
plt.show()
```

```python
# Survivors by gender
n_male_passengers=np.size(data[data[:,4]=='male',4])
n_male_survivors=np.sum(data[data[:,4]=='male',1].astype(np.int))
n_male_dead=n_male_passengers-n_male_survivors

n_female_passengers=np.size(data[data[:,4]=='female',4])
n_female_survivors=np.sum(data[data[:,4]=='female',1].astype(np.int))
n_female_dead=n_female_passengers-n_female_survivors

print('N. of MALE passengers:',n_male_passengers, '   || N. of MALE survivors:',
      n_male_survivors, '   || % of survivors:',100*n_male_survivors/n_passengers)
print('N. of FEMALE passengers:',n_female_passengers, ' || N. of FEMALE survivors:',
      n_female_survivors, ' || % of survivors:',100*n_female_survivors/n_passengers)
```

```python
# make a square figure and axes
plt.figure(1, figsize=(6,6))

# The slices will be ordered and plotted counter-clockwise.
labels = 'Male dead', 'Male survivors', 'Female survivors','Female dead'
fracs =[n_male_dead, n_male_survivors, n_female_survivors,n_female_dead]
explode=(0, 0.05, 0.05, 0)

plt.pie(fracs, explode=explode, labels=labels,
                autopct='%1.1f%%', shadow=True, startangle=90,
                colors=['blue', 'lightskyblue', 'lightcoral', 'red'])
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

plt.title('Passengers and survivors by gender', bbox={'facecolor':'0.8', 'pad':5})

plt.show()
```

```python
# Survivors by passenger class
n_1cl_passengers=np.size(data[data[:,2].astype(np.int)==1,2])
n_2cl_passengers=np.size(data[data[:,2].astype(np.int)==2,2])
n_3cl_passengers=np.size(data[data[:,2].astype(np.int)==3,2])

n_1cl_survivors=np.sum(data[data[:,2].astype(np.int)==1,1].astype(np.int))
n_2cl_survivors=np.sum(data[data[:,2].astype(np.int)==2,1].astype(np.int))
n_3cl_survivors=np.sum(data[data[:,2].astype(np.int)==3,1].astype(np.int))

n_1cl_dead=n_1cl_passengers-n_1cl_survivors
n_2cl_dead=n_2cl_passengers-n_2cl_survivors
n_3cl_dead=n_3cl_passengers-n_3cl_survivors

print('1st class passengers:',n_1cl_passengers,' || 1st class survivors:',n_1cl_survivors,' || %',100*n_1cl_survivors/n_1cl_passengers)
print('2nd class passengers:',n_2cl_passengers,' || 2nd class survivors:',n_2cl_survivors,'  || %',100*n_2cl_survivors/n_2cl_passengers)
print('3rd class passengers:',n_3cl_passengers,' || 3rd class survivors:',n_3cl_survivors,' || %',100*n_3cl_survivors/n_3cl_passengers)

```

```python
# make a square figure and axes
plt.figure(1, figsize=(13,6))
ax = plt.axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.
labels = '1st dead', '1st survivors', '2nd dead', '2nd survivors', '3rd dead', '3rd survivors'
fracs =[n_1cl_dead,n_1cl_survivors,n_2cl_dead,n_2cl_survivors,n_3cl_dead,n_3cl_survivors]
explode=(0,0.05,0,0.05,0, 0.05)

plt.subplot(121)
plt.pie(fracs, explode=explode, labels=labels,
                autopct='%1.1f%%', shadow=True, startangle=90,
                colors=['yellow','lightyellow', 'green','lightgreen','gray', 'lightgray'])
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

plt.title('Passengers and survivors by class', bbox={'facecolor':'0.8', 'pad':5})

# The slices will be ordered and plotted counter-clockwise.
labels2 = '1st survivors','2nd survivors','3rd survivors'
fracs2 =[n_1cl_survivors,n_2cl_survivors,n_3cl_survivors]
explode2=(0,0,0)

plt.subplot(122)
plt.pie(fracs2, explode=explode2, labels=labels2,
                autopct='%1.1f%%', shadow=True, startangle=90,
                colors=['lightyellow','lightgreen', 'lightgray'])
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

plt.title('Survivors by class', bbox={'facecolor':'0.8', 'pad':5})

plt.show()
```

```python
# 1st class by gender
n_1cl_survivors_m=np.sum(data[np.logical_and(data[:,2]=='1',data[:,4]=='male'),1].astype(np.int))
n_1cl_survivors_f=np.sum(data[np.logical_and(data[:,2]=='1',data[:,4]=='female'),1].astype(np.int))
n_1cl_dead_m=-np.sum(data[np.logical_and(data[:,2]=='1',data[:,4]=='male'),1].astype(np.int)-1)
n_1cl_dead_f=-np.sum(data[np.logical_and(data[:,2]=='1',data[:,4]=='female'),1].astype(np.int)-1)

# 1st class by gender
n_2cl_survivors_m=np.sum(data[np.logical_and(data[:,2]=='2',data[:,4]=='male'),1].astype(np.int))
n_2cl_survivors_f=np.sum(data[np.logical_and(data[:,2]=='2',data[:,4]=='female'),1].astype(np.int))
n_2cl_dead_m=-np.sum(data[np.logical_and(data[:,2]=='2',data[:,4]=='male'),1].astype(np.int)-1)
n_2cl_dead_f=-np.sum(data[np.logical_and(data[:,2]=='2',data[:,4]=='female'),1].astype(np.int)-1)

# 1st class by gender
n_3cl_survivors_m=np.sum(data[np.logical_and(data[:,2]=='3',data[:,4]=='male'),1].astype(np.int))
n_3cl_survivors_f=np.sum(data[np.logical_and(data[:,2]=='3',data[:,4]=='female'),1].astype(np.int))
n_3cl_dead_m=-np.sum(data[np.logical_and(data[:,2]=='3',data[:,4]=='male'),1].astype(np.int)-1)
n_3cl_dead_f=-np.sum(data[np.logical_and(data[:,2]=='3',data[:,4]=='female'),1].astype(np.int)-1)

# Plot fraction of survivors by gender and class
plt.figure(1, figsize=(13,6))
n_var = 3

index = np.arange(n_var)

height_cl_surv=np.array([n_1cl_survivors,n_2cl_survivors,n_3cl_survivors])
height_cl_dead=np.array([n_1cl_dead,n_2cl_dead,n_3cl_dead])

height_cl_surv_m=np.array([n_1cl_survivors_m,n_2cl_survivors_m,n_3cl_survivors_m])
height_cl_dead_m=np.array([n_1cl_dead_m,n_2cl_dead_m,n_3cl_dead_m])

height_cl_surv_f=np.array([n_1cl_survivors_f,n_2cl_survivors_f,n_3cl_survivors_f])
height_cl_dead_f=np.array([n_1cl_dead_f,n_2cl_dead_f,n_3cl_dead_f])

bar_width = 0.3

opacity = 1

plt.subplot(121)

rects_cl_surv = plt.bar(index, height_cl_surv, bar_width,
                 alpha=opacity,
                 color=('lightyellow','lightgreen', 'lightgray'))
rects_cl_dead = plt.bar(index, height_cl_dead, bar_width,
                 alpha=opacity,
                 color=('yellow','green', 'gray'),bottom=height_cl_surv)

rects_m_surv = plt.bar(index+bar_width, height_cl_surv_m, bar_width,
                 alpha=opacity,
                 color='lightblue')
rects_f_surv = plt.bar(index+bar_width, height_cl_surv_f, bar_width,
                 alpha=opacity,
                 color='lightcoral',bottom=height_cl_surv_m)

tmp=height_cl_surv_m+height_cl_surv_f
rects_m_dead = plt.bar(index+bar_width, height_cl_dead_m, bar_width,
                 alpha=opacity,
                 color='blue',bottom=tmp)
tmp=tmp+height_cl_dead_m
rects_f_dead = plt.bar(index+bar_width, height_cl_dead_f, bar_width,
                 alpha=opacity,
                 color='red',bottom=tmp)

plt.axis([-.5, 3.5, 0, 600])
plt.xlabel('variable')
plt.ylabel('Passengers')
plt.title('Survivors by class and gender')
plt.xticks(index + bar_width, ('1st','2nd','3rd'))

plt.grid(True)

plt.subplot(122)

n_cl_passengers=np.array([n_1cl_passengers,n_2cl_passengers,n_3cl_passengers])

rects_cl_surv = plt.bar(index, height_cl_surv/n_cl_passengers, bar_width,
                 alpha=opacity,
                 color=('lightyellow','lightgreen', 'lightgray'))
rects_cl_dead = plt.bar(index, height_cl_dead/n_cl_passengers, bar_width,
                 alpha=opacity,
                 color=('yellow','green', 'gray'),bottom=height_cl_surv/n_cl_passengers)

rects_m_surv = plt.bar(index+bar_width, height_cl_surv_m/n_cl_passengers, bar_width,
                 alpha=opacity,
                 color='lightblue')
rects_f_surv = plt.bar(index+bar_width, height_cl_surv_f/n_cl_passengers, bar_width,
                 alpha=opacity,
                 color='lightcoral',bottom=height_cl_surv_m/n_cl_passengers)

tmp=height_cl_surv_m+height_cl_surv_f
rects_m_dead = plt.bar(index+bar_width, height_cl_dead_m/n_cl_passengers, bar_width,
                 alpha=opacity,
                 color='blue',bottom=tmp/n_cl_passengers)
tmp=tmp+height_cl_dead_m
rects_f_dead = plt.bar(index+bar_width, height_cl_dead_f/n_cl_passengers, bar_width,
                 alpha=opacity,
                 color='red',bottom=tmp/n_cl_passengers)

plt.axis([-.5, 3.5, 0, 1.1])
plt.xlabel('variable')
plt.ylabel('Passengers (normalized to each class)')
plt.title('Survivors by class and gender (normalized)')
plt.xticks(index + bar_width, ('1st','2nd','3rd'))

plt.grid(True)


plt.show()
```

```python
# VARIABLE DESCRIPTIONS:
# 1  survival        Survival
#                    (0 = No; 1 = Yes)
# 2  pclass          Passenger Class
#                    (1 = 1st; 2 = 2nd; 3 = 3rd)
# 3  name            Name
# 4  sex             Sex
# 5  age             Age
# 6  sibsp           Number of Siblings/Spouses Aboard
# 7  parch           Number of Parents/Children Aboard
# 8  ticket          Ticket Number
# 9  fare            Passenger Fare
# 10 cabin           Cabin
# 11 embarked        Port of Embarkation
#                    (C = Cherbourg; Q = Queenstown; S = Southampton)
```

```python
%matplotlib inline

import numpy as np
import matplotlib as matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import pandas as pd
```

```python
train_path='./train.csv'
test_path='./test.csv'

# import data
traindf_raw = pd.read_csv(train_path)
testdf_raw = pd.read_csv(test_path)
```

```python
from clean import clean_data

# Clean the training values
traindf=clean_data(traindf_raw)
testdf=clean_data(testdf_raw)
```

```python
# Feature engineering

# Family size
traindf['FamSize']=traindf.Parch+traindf.SibSp
testdf['FamSize']=testdf.Parch+testdf.SibSp


# Log10Fare
traindf['Log10Fare']=traindf.Fare.apply(lambda x: np.log10(x))
testdf['Log10Fare']=testdf.Fare.apply(lambda x: np.log10(x))

# interaction between Age and Fare
#poly = preprocessing.PolynomialFeatures(2)
#xnew=poly.fit_transform(x)
```

```python
from sklearn import preprocessing

# Initialize the training vector
Xdf=pd.DataFrame()
X_test=pd.DataFrame()
featName=[]

# Pclass
tmp=pd.get_dummies(traindf['Pclass'])
tmp.columns=['pclass1','pclass2','pclass3']
Xdf = pd.concat([Xdf, tmp], axis=1)

tmp=pd.get_dummies(testdf['Pclass'])
tmp.columns=['pclass1','pclass2','pclass3']
X_test = pd.concat([X_test, tmp], axis=1)

# Sex
Xdf = pd.concat([Xdf, pd.get_dummies(traindf['Sex'])], axis=1)
X_test = pd.concat([X_test, pd.get_dummies(testdf['Sex'])], axis=1)

# Embarked
Xdf = pd.concat([Xdf, pd.get_dummies(traindf['Embarked'])], axis=1)
X_test = pd.concat([X_test, pd.get_dummies(testdf['Embarked'])], axis=1)

# SibSp
Xdf = pd.concat([Xdf, 1.*traindf['SibSp']/max(traindf['SibSp'])], axis=1)
X_test = pd.concat([X_test, 1.*testdf['SibSp']/max(traindf['SibSp'])], axis=1)

# Parch
Xdf = pd.concat([Xdf, 1.*traindf['Parch']/max(traindf['Parch'])], axis=1)
X_test = pd.concat([X_test, 1.*testdf['Parch']/max(traindf['Parch'])], axis=1)

# Family Size
Xdf = pd.concat([Xdf, 1.*traindf['FamSize']/max(traindf['FamSize'])], axis=1)
X_test = pd.concat([X_test, 1.*testdf['FamSize']/max(traindf['FamSize'])], axis=1)

# Age
tmp=preprocessing.scale(traindf['Age'])
tmp=pd.DataFrame({'Age':tmp})
Xdf = pd.concat([Xdf, tmp], axis=1)
tmp=preprocessing.scale(testdf['Age'])
tmp=pd.DataFrame({'Age':tmp})
X_test = pd.concat([X_test, tmp], axis=1)

# Log10Fare
tmp=preprocessing.scale(traindf['Log10Fare'])
tmp=pd.DataFrame({'Log10Fare':tmp})
Xdf = pd.concat([Xdf, tmp], axis=1)
tmp=preprocessing.scale(testdf['Log10Fare'])
tmp=pd.DataFrame({'Log10Fare':tmp})
X_test = pd.concat([X_test, tmp], axis=1)

# Fare-Age interaction
tmp=preprocessing.scale(traindf['Log10Fare']*traindf['Age'])
tmp=pd.DataFrame({'Fare-Age':tmp})
Xdf = pd.concat([Xdf, tmp], axis=1)
tmp=preprocessing.scale(testdf['Log10Fare']*testdf['Age'])
tmp=pd.DataFrame({'Fare-Age':tmp})
X_test = pd.concat([X_test, tmp], axis=1)

# Survivors
ydf=pd.DataFrame({'Survived':traindf['Survived']})
```

```python
featName=Xdf.columns
```

```python
# Convert to numpy array
X=np.squeeze(Xdf.as_matrix())
y=np.squeeze(ydf.as_matrix())
```

```python
# number of training examples and features
m,n=np.shape(X)
print('m =',m,'n=',n)
```

```python
# Null accuracy
null_accuracy=1-y.mean()

print('Null accuracy',null_accuracy)
```

```python
# Validation curve
from sklearn.learning_curve import validation_curve
from sklearn.svm import SVC

param_range=np.logspace(-5, 0, 20)

train_scores, test_scores = validation_curve(
    SVC(kernel='sigmoid'), X, y, param_name="gamma", param_range=param_range,
    cv=10, scoring="accuracy", n_jobs=1)

# Compute mean and std
train_scores_mean=np.mean(train_scores, axis=1)
train_scores_std=np.std(train_scores, axis=1)
test_scores_mean=np.mean(test_scores, axis=1)
test_scores_std=np.std(test_scores, axis=1)

# Plot validation curve
plt.figure(figsize=(10,6))
plt.title("Validation Curve with Logistic Regression")
plt.xlabel("C")
plt.ylabel("Score")
plt.ylim(0.5, 0.9)

plt.semilogx(param_range, train_scores_mean, label="Training score", color="r")
plt.fill_between(param_range, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.2, color="r")
plt.semilogx(param_range, test_scores_mean, label="Cv score", color="g")
plt.fill_between(param_range, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.2, color="g")
plt.legend(loc="best")
plt.grid()
plt.show()
```

```python
# Try learning curve
from sklearn.learning_curve import learning_curve

param_sizes=np.arange(0.05,1.05,.05)

train_sizes, train_scores, valid_scores = learning_curve(
    SVC(gamma=0.1,kernel='poly',degree=3), X, y, train_sizes=param_sizes,cv=10, scoring='accuracy')

# Compute mean and std
train_scores_mean=np.mean(train_scores, axis=1)
train_scores_std=np.std(train_scores, axis=1)
valid_scores_mean=np.mean(valid_scores, axis=1)
valid_scores_std=np.std(valid_scores, axis=1)

# Plot learning curve
plt.figure(figsize=(10,6))
plt.title("Learnig Curve with Logistic Regression C=1")
plt.xlabel("training examples")
plt.ylabel("Score")
plt.ylim(0.6, 1)

plt.plot(train_sizes, train_scores_mean, label="Training score", color="r")
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.2, color="r")
plt.plot(train_sizes, valid_scores_mean, label="Cv score", color="g")
plt.fill_between(train_sizes, valid_scores_mean - valid_scores_std,
                 valid_scores_mean + valid_scores_std, alpha=0.2, color="g")
plt.legend(loc="best")
plt.grid()
plt.show()
```

```python
# Define logistic regression object
svm=SVC(gamma=0.1,probability=True,kernel='poly',degree=3)
svm.fit(X,y)
```

```python
# Try confusion matrix
from sklearn.metrics import confusion_matrix
import sklearn.metrics as metrics
from sklearn.preprocessing import binarize

y_pred=svm.predict(X)

conf=confusion_matrix(y,y_pred)
print('Confusion matrix')
print(conf)


TP=conf[1,1]
TN=conf[0,0]
FP=conf[0,1]
FN=conf[1,0]

print('NullAccuracy:                     ',null_accuracy)
print('Accuracy   : (TP+TN)/(TP+TN+FP+FN)',(TP+TN)/(TP+TN+FP+FN),metrics.accuracy_score(y,y_pred))
print('Class Error: (FP+FN)/(TP+TN+FP+FN)',(FP+FN)/(TP+TN+FP+FN),1-metrics.accuracy_score(y,y_pred))
print('Sensitivity:      TP/(TP+FN)      ',TP/(TP+FN),metrics.recall_score(y,y_pred))
print('Specificity:      TN/(TN+FP)      ',TN/(TN+FP))
print('FP rate    :      FP/(FP+TN)      ',FP/(FP+TN))
print('Precision  :      TP/(FP+TP)      ', TP/(FP+TP),metrics.precision_score(y,y_pred) )
print('F1 score   :                      ', metrics.f1_score(y,y_pred) )
```

```python
# Hist of the probability threshold
y_pred_prob=svm.predict_proba(X)[:,1]

# Plot learning curve
plt.figure(figsize=(10,6))
plt.title("Predicted probability histogram")
plt.xlabel("predicted probability")
plt.ylabel("counts")
plt.xlim(0,1)
plt.hist(y_pred_prob,bins=10, color="b",alpha=0.2)
plt.show()
```

```python
# Try ROC curve

fpr, tpr, thresholds = metrics.roc_curve(y, y_pred_prob)

acc=[]
for i in range(0,len(thresholds)):
    y_pred_mod=binarize(y_pred_prob,thresholds[i])[0]
    acc.append(metrics.accuracy_score(y,y_pred_mod))
```

```python
# Plot ROC curve
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
plt.title("ROC curve")
plt.xlabel("FP rate (1-specificity)")
plt.ylabel("TP rate (sensitivity)")
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.grid()
plt.plot(fpr,tpr, color="b")
plt.plot([0, 1], [0, 1], 'k--')

plt.subplot(1,2,2)
plt.title("FP/TP curves")
plt.xlabel("thresholds")
plt.ylabel("rate")
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.plot(thresholds,tpr, color="r",label="TP")
plt.plot(thresholds,fpr, color="g",label="FP")
plt.plot(thresholds,acc, color="k",label="acc")
plt.legend(loc="best")
plt.grid()

plt.show()

metrics.roc_auc_score(y, y_pred_prob)
```

```python
# Predict
y_pred_prob_test=svm.predict_proba(X_test)[:,1]
y_pred_test=binarize(y_pred_prob_test,0.2)[0].astype(int)
```
